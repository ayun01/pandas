```python
import pandas as pd
import numpy as np
```

```python
titanic_survival = pd.read_csv("titanic_train.csv")
titanic_survival.head()
```

```python
age = titanic_survival["Age"]
age_is_null = pd.isnull(age)
age_null = age[age_is_null]
age_null_count = len(age_null)
print(age_null)
print(age_null_count)
```

```python
# fill null with mean of age
age_has_value = titanic_survival["Age"][age_is_null == False]
mean_age = sum(age_has_value) / len(age_has_value)
print(mean_age)
# or
mean_age = titanic_survival["Age"].mean()
print(mean_age)
```

```python
#mean fare for each class
passenger_classes = [1, 2, 3]
fare_by_passenger_class = {}
for cls in passenger_classes:
    psgclass_fare = titanic_survival["Fare"][titanic_survival["Pclass"] == cls]
    fare_mean = psgclass_fare.mean()
    fare_by_passenger_class[cls] = fare_mean
print(fare_by_passenger_class)
```

# pivot_table
** relation between 1 column and other columns **

```python
# pivot_table
passenger_survival = titanic_survival.pivot_table(index='Pclass',values='Survived',aggfunc=np.mean)
print(passenger_survival)
```

```python
# pivot_table
# by default, aggfun = mean
passenger_survival = titanic_survival.pivot_table(index='Embarked',values=['Fare','Survived'],aggfunc=np.sum)
print(passenger_survival)
```

```python
drop_na_columns = titanic_survival.dropna(axis=1) # 1 = column
new_titanic_survival = titanic_survival.dropna(axis=0,subset=['Age','Sex'])
print(new_titanic_survival)
```

# custom function

```python
# the one hundredth element
def hundredth_row(column):
    hundredth_item = column.loc[99]
    return hundredth_item

hundredth_row = titanic_survival.apply(hundredth_row)
print(hundredth_row)
```

```python
# count null value for each column
def null_count(column):
    column_null = pd.isnull(column)
    null = column[column_null]
    return len(null)

col_null_count = titanic_survival.apply(null_count)
print(col_null_count)
```

```python
# bin age
def is_minor(row):
    if row['Age'] < 18:
        return True
    else:
        return False

minors = titanic_survival.apply(is_minor, axis=1)
def generate_age_label(row):
    age = row['Age']
    if pd.isnull(age):
        return 'Unknown'
    elif age < 18:
        return 'Minor'
    else:
        return 'Adult'
age_label = titanic_survival.apply(generate_age_label,axis=1)
print(age_label)
```

```python
titanic_survival['age_label'] = age_label
survival_by_age = titanic_survival.pivot_table(index='age_label',values='Survived')
print(survival_by_age)
```
